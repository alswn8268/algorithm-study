## 문제 요약

> 정수로 된 배열(또는 리스트)에서 합이 50이 되는 두 수를 찾아라

## 접근 방식

- 아이디어 설명
- 알고리즘 선택 이유

```python
# 풀이 코드
arr = [10, 20, 30, 40, 50, 60, 70] # input 배열
obs = 50 # 목표 합

left = 0 # 왼쪽 포인터
right = len(arr) - 1 # 오른쪽 포인터 

while (left < right): # 두 포인터가 만날 때까지 반복
    res = arr[left] + arr[right]
    if (res == obs):
        break
    elif (res < obs):
        left += 1
    else:
        right -= 1

print(res, arr[left], arr[right])
```

## 복잡도 분석

- **시간 복잡도**: O( )
- **공간 복잡도**: O( )
