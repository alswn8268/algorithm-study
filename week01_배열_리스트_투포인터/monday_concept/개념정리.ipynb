## 번외) 시간 복잡도란?

> 입력 크기(n)가 커질수록 실행 시간이 어떻게 늘어나는지를 나타낸 것  
> O(x)은 최악 시간 복잡도

```python
# 예시 1) 하나씩 찾기 → O(n)
# 배열을 처음부터 끝까지 확인 → 데이터가 n개면 최대 n번 반복
for x in arr:
    if x == target:
        print(True)
```

```python
# 예시 2) 바로 접근 → O(1)
# 한 번에 바로 접근
arr = [1, 2, 3, 4]
print(arr[3])
```

```python
# 예시 3) 이중 반복 → O(n²)
arr = [1, 2, 3, 4]

for i in arr:
    for j in arr:
        print(i, j)
```

### 왜 중요한가?

→ 데이터가 작을 때는 별 차이 없지만, 데이터가 커지거나 양이 많아지면 차이가 폭발적으로 커지기 때문

## 배열 (Array)

> 메모리에 연속 공간에 값이 채워져 있는 형태의 자료 구조

- 인덱스를 통해 참조할 수 있음
- 선언한 자료형의 값만 저장할 수 있음
- 새로운 값을 삽입하거나 특정 인덱스에 있는 값을 삭제하기 어려움  
  (값을 삽입하거나 삭제하려면 해당 인덱스 주변에 있는 값을 이동시키는 과정이 필요)
- 배열의 크기는 선언할 때 지정할 수 있으며, 한 번 선언하면 크기를 늘리거나 줄일 수 없음

| 연산 | 시간복잡도 |
| --- | --- |
| 접근 (arr[i]) | O(1) |
| 수정 | O(1) |
| 탐색 (값 찾기) | O(n) |
| 삽입 (중간) | O(n) |
| 삭제 (중간) | O(n) |

**비유**: 책꽂이에 번호 붙은 책  
- 3번 책 → 바로 꺼냄 (빠름)  
- 중간에 책 넣기 → 자리 밀어야 함 (느림)

**예시**: 시험 점수 목록 / 좌표 리스트 / 이미지 픽셀 데이터

> 💡 파이썬에서는 배열과 리스트를 구분하지 않음!

```python
# Python - 배열처럼 사용하는 list
arr = [10, 20, 30, 40]

print(arr[2])       # 인덱스로 바로 접근 (O(1)) → 30

arr[1] = 25         # 값 수정 (index 1 → 25)

arr.insert(2, 99)   # index 2 위치에 99 삽입 (뒤 요소들이 밀림 → O(n))

arr.pop(0)          # index 0 요소 삭제 (뒤가 앞으로 당겨짐)

print(arr)          # [25, 99, 30, 40]
```

## 리스트 (Linked List)

> 노드(값과 포인터를 묶은 것)를 포인터로 연결한 자료 구조

- 인덱스가 없으므로 값에 접근하려면 Head 포인터부터 순서대로 접근해야 함 → 접근 속도가 느림
- 포인터로 연결되어 있으므로 데이터를 삽입하거나 삭제하는 연산 속도가 빠름
- 선언할 때 크기를 별도로 지정하지 않아도 됨
- 포인터를 저장할 공간이 필요하므로 배열보다 구조가 복잡함

| 연산 | 시간복잡도 |
| --- | --- |
| 접근 (k번째) | O(n) |
| 탐색 | O(n) |
| 삽입 (노드 알고 있을 때) | O(1) |
| 삭제 (노드 알고 있을 때) | O(1) |

**비유**: 보물찾기 지도  
- "다음 위치는 여기야!" 하고 계속 따라감  
- 바로 점프는 못 함

**예시**: 음악 플레이리스트 / Undo·Redo 시스템 / 메모리 효율이 중요한 경우

```python
# Python - 연결 리스트 직접 구현
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None   # 다음 노드를 가리키는 포인터

# 연결 리스트 생성
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

# 순회
current = head
while current:
    print(current.data)    # 10 → 20 → 30
    current = current.next
```

## 투 포인터 (Two Pointers)

> 2개의 포인터로 알고리즘의 시간 복잡도를 최소화

- 시간복잡도: **O(n)** (각 포인터는 최대 n번만 움직이기 때문)
- 정렬된 배열에서 매우 강력함

**비유**: 양쪽에서 좁혀오는 탐색  
- 목표에 따라 한쪽을 줄이면서 찾음

**예시**: 합이 50이 되는 두 수 찾기

```python
# Python - 투 포인터
arr = [10, 20, 30, 40, 50]
target = 50

left = 0
right = len(arr) - 1

while left < right:
    s = arr[left] + arr[right]

    if s == target:
        print(arr[left], arr[right])  # 10 40
        break
    elif s < target:   # 합이 작으면 → 왼쪽 증가 (더 크게)
        left += 1
    else:              # 합이 크면  → 오른쪽 감소 (더 작게)
        right -= 1
```
