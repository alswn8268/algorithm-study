## 3. 투포인터 (Two Pointers)

- 언제 쓰나: 정렬된 배열에서 두 원소의 합/차 조건을 만족하는 쌍 탐색
- 시간 복잡도: O(n)

```python
# 투포인터 기본 패턴
def two_pointer(arr, target):
    left, right = 0, len(arr) - 1
    while left < right:
        s = arr[left] + arr[right]
        if s == target:
            return [left, right]
        elif s < target:
            left += 1
        else:
            right -= 1
    return []
```
